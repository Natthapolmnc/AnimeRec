```
###Load data
rating=pd.read_csv("drive/Shared drives/Chula Drive/Datasets/AnimeRec-dataset/rating.csv")
anime=pd.read_csv("drive/Shared drives/Chula Drive/Datasets/AnimeRec-dataset/anime.csv")
```

```
###Data Preprocessing
rating.rating=np.logical_or(rating.rating==-1,rating.rating>=4).astype(int)
anime=anime[anime.type=="Movie"]
anime=anime.drop(columns=["genre","type","episodes","rating","members"])
rating=rating[:100000].merge(anime,left_on="anime_id",right_on="anime_id")
impli_matrix=pd.pivot_table(rating,index="user_id",columns="name",values="rating",aggfunc=np.sum,fill_value=0)
impli_matrix.to_csv("drive/Shared drives/Chula Drive/Datasets/AnimeRec-dataset/impli_matrix.csv")
impli_matrix
```
